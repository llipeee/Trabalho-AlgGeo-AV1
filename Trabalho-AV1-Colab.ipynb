## 1) Verificação do projeto
Esta célula confere se os arquivos principais do trabalho estão presentes no mesmo diretório do notebook.


````python
from pathlib import Path
import os
import sys

base_dir = Path.cwd()
required_files = [
    'main.py',
    'tests.py',
    'matrix.py',
    'vector.py',
    'linear_algebra.py',
    'README.md',
]

print('Diretório atual:', base_dir)
missing = [name for name in required_files if not (base_dir / name).exists()]

if missing:
    print('
Arquivos ausentes:')
    for name in missing:
        print('-', name)
    raise FileNotFoundError(
        'O notebook precisa estar na raiz do projeto, junto dos arquivos Python.'
    )

print('
Todos os arquivos principais foram encontrados com sucesso.')
print('
Conteúdo da pasta:')
for item in sorted(base_dir.iterdir()):
    if item.name.startswith('.'):  # evita poluição visual
        continue
    marker = '/' if item.is_dir() else ''
    print('-', item.name + marker)
````

## 2) Ajuste do ambiente
Adiciona o diretório atual ao `sys.path` para garantir que os imports funcionem no Colab.


````python
if str(base_dir) not in sys.path:
    sys.path.append(str(base_dir))

print('sys.path ajustado com sucesso.')
````

## 6) Exemplos adicionais de uso
Esta célula permite demonstrar rapidamente o uso das classes principais durante apresentação ou teste no Colab.


````python
from matrix import Matrix
from vector import Vector
from linear_algebra import LinearAlgebra

A = Matrix(2, 2, [1, 2, 3, 4])
B = Matrix(2, 2, [10, 20, 30, 40])
v = Vector(2, [5, 6])

print('Matriz A:')
print(A)
print('
Matriz B:')
print(B)
print('
Vetor v:')
print(v)
print('
A + B:')
print(LinearAlgebra.sum(A, B))
print('
2 * v:')
print(LinearAlgebra.times(2, v))
print('
Transposta de A:')
print(LinearAlgebra.transpose(A))
````
